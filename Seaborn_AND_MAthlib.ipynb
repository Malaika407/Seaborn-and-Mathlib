```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

import pandas as pd


```

```python
df=pd.read_csv("Iris.csv")
```

```python
df
```

```python
df.corr()
```

```python
x = df['Id']
y = df['SepalLengthCm']
plt.plot(x, y)
plt.xlabel('Id')
plt.ylabel('SepalLengthCm')

plt.show()
```

```python
plt.scatter(x,y,c='b')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Graph in 2D')
plt.savefig('Test.png')
```

```python

plt.plot(x,y,'b',linestyle='dashed',linewidth=1, markersize=10)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('2d Diagram')
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(x,y,'g*--')
plt.subplot(2,2,3)
plt.plot(x,y,'bo')
plt.subplot(2,2,4)
plt.plot(x,y,'g--')
```

```python
# Data to plot
labels = 'Id	', 'SepalLengthCm', 'size'
sizes = [215, 130, 245]
colors = ['gold', 'yellowgreen', 'lightcoral']
explode = (0.6, 0, 0)  # explode 1st slice

# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True)

plt.axis('equal')
plt.show()
```

```python
x = df['SepalLengthCm']
plt.hist(x ,bins = 10)
plt.title("histogram")
plt.show()
```

```python
## Bar plot

x = df['Id']
y = df['SepalLengthCm']

plt.bar(x, y, color='g')

plt.title('Bar graph')
plt.ylabel('Y axis')
plt.xlabel('X axis')

plt.show()
```
